Chunk the original pdf

````python
import os
import pdfplumber

# === CONFIGURATION ===
input_pdf_folder = "Original pdf/layer3_v11"
output_base_folder = "acts/layer3_v11"
max_tokens = 3000  # Approximate word limit per chunk

# === FUNCTION: Extract text from PDF ===
def extract_text_from_pdf(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        text = ""
        for page in pdf.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text + "\n"
    return text

# === FUNCTION: Chunk text ===
def chunk_text(text, max_tokens=3000):
    chunks = []
    current_chunk = []
    current_length = 0

    paragraphs = [p.strip() for p in text.split("\n") if p.strip()]

    for paragraph in paragraphs:
        paragraph_length = len(paragraph.split())
        if paragraph_length > max_tokens:
            words = paragraph.split()
            for i in range(0, len(words), max_tokens):
                sub_paragraph = " ".join(words[i:i + max_tokens])
                chunks.append(sub_paragraph)
        elif current_length + paragraph_length > max_tokens:
            chunks.append("\n\n".join(current_chunk))
            current_chunk = []
            current_length = 0
        current_chunk.append(paragraph)
        current_length += paragraph_length

    if current_chunk:
        chunks.append("\n\n".join(current_chunk))

    return chunks

# === FUNCTION: Save chunks ===
def save_chunks_to_txt(chunks, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    for i, chunk in enumerate(chunks):
        chunk_file_path = os.path.join(output_dir, f"chunk_{i + 1}.txt")
        with open(chunk_file_path, "w", encoding="utf-8") as f:
            f.write(chunk)
    print(f"âœ… Saved {len(chunks)} chunks to '{output_dir}'")

# === MAIN PROCESS ===
for filename in os.listdir(input_pdf_folder):
    if filename.lower().endswith(".pdf"):
        act_name = filename.replace(".pdf", "").strip()
        input_path = os.path.join(input_pdf_folder, filename)
        output_dir = os.path.join(output_base_folder, act_name)

        print(f"\nðŸ“„ Processing: {act_name}")
        raw_text = extract_text_from_pdf(input_path)
        text_chunks = chunk_text(raw_text, max_tokens=max_tokens)
        save_chunks_to_txt(text_chunks, output_dir)

print("\nðŸŽ‰ All PDFs have been processed and chunked.")
````

Function to call Claud 3 Haiku Pipeline

````python
import re
import json
import anthropic
import os

if not client.api_key:
    raise ValueError("ðŸš¨ ANTHROPIC_API_KEY is not set! Configure it before running the script.")

def extract_acts_relationships_claude(chunk_text, source_act):
    
    prompt = f"""
    Analyze the following legal text and extract all **external** acts and their relationships.

    **Rules:**
    - Only extract **external references** (ignore internal references to the same act).
    - Standardize act names (e.g., "1995 Act" â†’ "Domestic Violence Act 1995").
    - Remove duplicate entries and ensure consistency.

    **Valid Relationship Types:**
    - "Amended" â†’ This act modifies another act.
    - "Repealed" â†’ This act removes another act.
    - "Referenced" â†’ This act only mentions another act but does not modify it.

    **Return the output in strict JSON format:**
    ```json
    {{
      "acts": [
        {{"source": "{source_act}", "target": "Sentencing Act 2002", "relationship": "Referenced"}},
        {{"source": "{source_act}", "target": "Privacy Act 2020", "relationship": "Amended"}}
      ]
    }}
    ```

    Legal Text:
    {chunk_text}
    """

    try:
        response = client.messages.create(
            model="claude-3-haiku-20240307",
            max_tokens=3000,
            temperature=0.2,
            messages=[
                {"role": "user", "content": prompt}
            ]
        )

        raw_response = response.content[0].text.strip()
        print(f"ðŸ“ Raw API Response:\n{raw_response[:400]}...")

        # âœ… Extract JSON part using regex
        match = re.search(r"\{[\s\S]*\}", raw_response)
        if match:
            json_text = match.group(0)
            return json.loads(json_text)
        else:
            print("âš ï¸ No JSON block found in response.")
            return {"acts": []}

    except Exception as e:
        print(f"ðŸš¨ Error calling Claude or parsing response: {e}")
        return {"acts": []}
````

````python
# === MAIN AUTOMATION ===

base_folder = "acts/layer3_v11"
output_file = "layer_2_relationships_appended.json"
all_results = []

# Loop through each act folder
for act_name in sorted(os.listdir(base_folder)):
    act_folder_path = os.path.join(base_folder, act_name)
    if not os.path.isdir(act_folder_path):
        continue

    print(f"\nProcessing act: {act_name}")

    # Loop through each chunk
    for file in sorted(os.listdir(act_folder_path)):
        if file.endswith(".txt"):
            with open(os.path.join(act_folder_path, file), "r", encoding="utf-8") as f:
                text = f.read().strip()
                if text:
                    result = extract_acts_relationships_claude(text, source_act=act_name)
                    if result:
                        all_results.append(result)

# Load existing results
if os.path.exists(output_file):
    with open(output_file, "r", encoding="utf-8") as f:
        existing_data = json.load(f)
else:
    existing_data = []

# Combine and save
combined_data = existing_data + all_results

with open(output_file, "w", encoding="utf-8") as f:
    json.dump(combined_data, f, indent=2)

print(f"\nâœ… All relationships saved to: {output_file}")
````

Clean the json output

````python
import json

# Load the original file
with open("layer_2_relationships_appended.json", "r", encoding="utf-8") as f:
    raw_data = json.load(f)

flat_relationships = []

# Flatten all "acts" entries into one list
for entry in raw_data:
    acts = entry.get("acts", [])
    flat_relationships.extend(acts)

# Remove duplicates based on (source, target, relationship)
unique_relationships = []
seen = set()

for item in flat_relationships:
    # Skip incomplete entries
    if not all(k in item for k in ("source", "target", "relationship")):
        print(f"âš ï¸ Skipped incomplete entry: {item}")
        continue

    key = (item["source"], item["target"], item["relationship"])
    if key not in seen:
        seen.add(key)
        unique_relationships.append(item)

# Save cleaned list to a new file
with open("layer_2_relationships_cleaned.json", "w", encoding="utf-8") as f:
    json.dump(unique_relationships, f, indent=2)

print(f"Cleaned and deduplicated file saved to 'layer_2_relationships_cleaned.json'")
print(f"Total unique relationships: {len(unique_relationships)}")
````

Get a count of acts

````python
from collections import defaultdict

# Load the cleaned file
with open("layer_2_relationships_cleaned.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Count total relationships
total_relationships = len(data)

# Track all unique act names (source or target)
unique_acts = set()

# Track number of relationships per act (as source OR target)
act_relationship_count = defaultdict(int)

for entry in data:
    source = entry["source"]
    target = entry["target"]

    unique_acts.add(source)
    unique_acts.add(target)

    act_relationship_count[source] += 1
    act_relationship_count[target] += 1

# Print overall stats
print(f"Total relationships: {total_relationships}")
print(f"Unique acts involved (source or target): {len(unique_acts)}")
print(f"\nNumber of relationships per act (source or target):")

for act, count in sorted(act_relationship_count.items(), key=lambda x: x[1], reverse=True):
    print(f"  - {act}: {count}")
````

Find the newly added acts

````python
# Load the cleaned relationship data
with open("layer_2_relationships_cleaned.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Step 1: Collect all unique acts in the network (source + target)
unique_acts = set()
for entry in data:
    source = entry.get("source")
    target = entry.get("target")
    if source: unique_acts.add(source)
    if target: unique_acts.add(target)

# Step 2: Get list of chunked act folders under acts/
chunked_act_folders = set(os.listdir("acts"))

# Step 3: Find acts that are in the network but not chunked
not_chunked_acts = sorted([act for act in unique_acts if act not in chunked_act_folders])

# Step 4: Output
print(f"ðŸ“„ Total acts in the network: {len(unique_acts)}")
print(f"ðŸ“ Total acts already chunked: {len(chunked_act_folders)}")
print(f"âŒ Acts in network but NOT chunked: {len(not_chunked_acts)}")
print("\nMissing act folders:")
for act in not_chunked_acts:
    print(f"- {act}")

# Optional: save to a file
with open("missing_acts.txt", "w", encoding="utf-8") as f:
    for act in not_chunked_acts:
        f.write(act + "\n")
print("âœ… Missing acts list saved to 'missing_acts.txt'")
````
